# Adquisicion de Data

```python
import pandas as pd

df_movies = pd.read_csv(r'.\Data\rotten_tomatoes_movies.csv')
df_reviews = pd.read_csv(r'.\Data\rotten_tomatoes_critic_reviews.csv')


```

Creamos el corpus con las columnas que creemos que son necesarias 

```python

#merged_data = pd.merge(df_movies,df_reviews, on='rotten_tomatoes_link', how='inner')
df_corpus = pd.merge( df_movies[[ 'rotten_tomatoes_link','movie_title']],df_reviews[['rotten_tomatoes_link','review_content']],  on='rotten_tomatoes_link')
```

### Convertir a minusculas y eliminar caracteres especiales

```python
import pandas as pd
import re

def limpiarTexto(texto):
    # Convertir a minÃºsculas
    texto = texto.lower()
    # Eliminar caracteres especiales (excepto espacios y letras)
    texto = re.sub(r'[^a-zA-Z\s]', '', texto)
    return texto
```

```python

df_corpus['textoLimpio'] = df_corpus['review_content'].apply(limpiarTexto)
df_corpus
```

```python
# El error de anterior nos indica que en la columna review se encuentran datos Nan que nos impide procesar los textos
# Los eliminamos con dropna
df_corpus.dropna(subset=['review_content'], inplace=True)
df_corpus.head()
```

```python
# Procedemos a limpiar el texto y creamos una columna en el dataframe corpus
df_corpus['textoLimpio'] = df_corpus['review_content'].apply(limpiarTexto)
df_corpus.head()
```

### EliminaciÃ³n de stopwords y tokenizacion

```python

import pandas as pd
from nltk.corpus import stopwords
import nltk

# Descargar stopwords de NLTK
nltk.download('stopwords')

# FunciÃ³n para eliminar stopwords
def eliminarStopwords(texto):
    palabras = texto.split()  # Divide el texto en palabras
    stop_words = set(stopwords.words('english'))  # ObtÃ©n las stopwords en inglÃ©s
    palabras_filtradas = [palabra for palabra in palabras if palabra.lower() not in stop_words]  # Filtra las stopwords
    return palabras_filtradas  #Recosntruye el texto sin las stopwords
```

```python
df_corpus['textoSinStopWordsTokenizado'] = df_corpus['textoLimpio'].apply(eliminarStopwords)
```

```python
df_corpus.head()
```

### Lemmatizacion

```python
import nltk
from nltk.stem import WordNetLemmatizer

# Descargar los recursos necesarios de NLTK
nltk.download('wordnet')

# Inicializar el lemmatizer de NLTK
#lemmatizer = WordNetLemmatizer()

# FunciÃ³n para lematizar el texto
def lematizarTexto(lista_palabras):
    lemmatizer = WordNetLemmatizer()
    palabrasLematizadas = [lemmatizer.lemmatize(palabra) for palabra in lista_palabras]  # Lematiza cada palabra
    return ' '.join(palabrasLematizadas)
```

```python
# Aplicar la lematizaciÃ³n a la columna de texto
df_corpus['textoLematizado'] = df_corpus['textoSinStopWordsTokenizado'].apply(lematizarTexto)
```

# Construccion del Sistema

### VectorizaciÃ³n utilizando tf-idf

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import pandas as pd

# VectorizaciÃ³n utilizando TF-IDF
vectorizer = TfidfVectorizer()
# Aplicar la vectorizaciÃ³n sobre el corpus con el texto preprocesado
X_corpus = vectorizer.fit_transform(df_corpus['textoLematizado'])
```

### Vectorizacion de la Query

```python
# Vectorizar la query ene el mismo espacio vectorial del corpus
query1 = "Movies about space travel"
X_query1 = vectorizer.transform([query1])
```

```python
# Vectorizar la query ene el mismo espacio vectorial del corpus
query2 = "Movies to watch with the family"
X_query2 = vectorizer.transform([query2])
```

### Calculo de la similitud entre la query y el corpus

```python
import numpy as np

# Calcular la similitud de coseno entre la query y el corpus
cosineSimilarityQuery1 = cosine_similarity(X_query1, X_corpus)

# Ordenar las similitudes de mayor a menor y obtener los Ã­ndices de los 10 documentos mÃ¡s similares
topQuery1 = np.argsort(cosineSimilarityQuery1[0])[::-1][:10]
```

```python

# Calcular la similitud de coseno entre la query y el corpus
cosineSimilarityQuery2 = cosine_similarity(X_query2, X_corpus)

# Ordenar las similitudes de mayor a menor y obtener los Ã­ndices de los 10 documentos mÃ¡s similares
topQuery2 = np.argsort(cosineSimilarityQuery2[0])[::-1][:10]
```

### PresentaciÃ³n de los resultados del sistema

```python
print(f"Query 1: {query1} \n")
i = 1 
#En este bucle recorremos los 10 documentos y recuperamos la informacion de dichos indices
for idx in topQuery1 :
    similitud = cosineSimilarityQuery1[0][idx]  # Grado de similitud para el documento
    print(f"Top {i}")
    print(f"Documento {idx}:")
    print(f"  Grado de similitud: {similitud:.4f}")
    print(f"  TÃ­tulo: {df_corpus['movie_title'].iloc[idx]}")
    print(f"  ReseÃ±a: {df_corpus['review_content'].iloc[idx]}")
    print("\n")
    i = i+1
    
```

```python
print(f"Query 2: {query2} \n")
i = 1 
for idx in topQuery2:
    print(f"Top {i}")
    similitud = cosineSimilarityQuery2[0][idx]  # Grado de similitud para el documento
    print(f"Documento {idx}:")
    print(f"  Grado de similitud: {similitud:.4f}")
    print(f"  TÃ­tulo: {df_corpus['movie_title'].iloc[idx]}")
    print(f"  ReseÃ±a: {df_corpus['review_content'].iloc[idx]}")
    print("\n")
    i = i+1
    
```
