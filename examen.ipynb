```python
import pandas as pd
import re


# FunciÃ³n de preprocesamiento
def preprocesar_texto(texto):
    # Convertir a minÃºsculas
    texto = texto.lower()
    # Eliminar caracteres especiales (excepto espacios y letras)
    texto = re.sub(r'[^a-zA-Z\s]', '', texto)
    return texto
```

```python
#Eliminar valores Nan y 
df_corpus.dropna(subset=['review_content'], inplace=True)
df_corpus
```

```python
#Error generado por los Nan de la columna review

df_corpus['textoLimpio'] = df_corpus['review_content'].apply(preprocesar_texto)
df_corpus
```

```python
import pandas as pd
from nltk.corpus import stopwords
import nltk

# Descargar stopwords de NLTK
nltk.download('stopwords')

# FunciÃ³n para eliminar stopwords
def eliminarStopwords(texto):
    palabras = texto.split()  # Divide el texto en palabras
    stop_words = set(stopwords.words('english'))  # ObtÃ©n las stopwords en inglÃ©s
    palabras_filtradas = [palabra for palabra in palabras if palabra.lower() not in stop_words]  # Filtra las stopwords
    return ' '.join(palabras_filtradas)

  # Devuelve la lista de palabras filtradas
```

```python
#La data de reviews es muy grande :/ Pero pues no puedo acortarla porque se supone que debe devolverme los resultados 
```

```python
df_corpus['textoSinStopWords'] = df_corpus['textoLimpio'].apply(eliminarStopwords)
```

```python
df_corpus
```

```python
#Demasiada data para que una humilde laptop la pueda procesar inge asi que limitamos las review a unicamente tres :)
df_corpus = df_corpus.groupby('movie_title').head(3).reset_index(drop=True)
```

```python
import nltk
from nltk.stem import WordNetLemmatizer

# Descargar los recursos necesarios de NLTK
nltk.download('wordnet')

# Inicializar el lemmatizer de NLTK
lemmatizer = WordNetLemmatizer()

# FunciÃ³n para lematizar el texto
def lematizar_texto(texto):
    # Dividir el texto en palabras
    palabras = texto.split()
    
    # Lematizar cada palabra
    palabras_lematizadas = [lemmatizer.lemmatize(palabra) for palabra in palabras]
    
    # Devolver el texto lematizado como una cadena
    return ' '.join(palabras_lematizadas)
```

```python
# Aplicar la lematizaciÃ³n a la columna de texto
df_corpus['textoLematizado'] = df_corpus['textoSinStopWords'].apply(lematizar_texto)
```

# Construccion del Sistema

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# Inicializamos el vectorizador TF-IDF
vectorizer = TfidfVectorizer(stop_words=None, max_features=5000)  # 'stop_words=None' porque ya eliminaste las stopwords

# Aplicamos el vectorizador a la columna 'textoSinStopWords'
X = vectorizer.fit_transform(df_corpus['textoLematizado'])

# Ver el tamaÃ±o de la matriz
print(X.shape)  # Esto te dirÃ¡ cuÃ¡ntas filas (documentos) y columnas (palabras Ãºnicas) tienes

# Convertimos la matriz dispersa a un DataFrame para ver las caracterÃ­sticas
df_tfidf = pd.DataFrame(X.toarray(), columns=vectorizer.get_feature_names_out())

print(df_tfidf.head())

```

```python
# Queries de ejemplo
queries = ["Movies about space travel", "Movies to watch with family"]

# Preprocesamiento de las queries (limpieza)
queries_cleaned = [preprocesar_texto(query) for query in queries]  # Usamos la funciÃ³n de limpieza que ya tienes

# VectorizaciÃ³n con el mismo TfidfVectorizer que usaste para el corpus
queries_vectorized = vectorizer.transform(queries_cleaned)

# Convertimos la matriz dispersa a un DataFrame para ver las caracterÃ­sticas
queries_df = pd.DataFrame(queries_vectorized.toarray(), columns=vectorizer.get_feature_names_out())

# Ver el resultado
print(queries_df)
```

```python
# Lista de queries
queries = ["Movies about space travel", "Movies to watch with family"]

# Preprocesamiento de las queries (usando la misma funciÃ³n de preprocesamiento que definiste)
queries_cleaned = [preprocesar_texto(query) for query in queries]
queries_cleaned = [eliminarStopwords(query) for query in queries]
queries_cleaned = [lematizar_texto(query) for query in queries]
# Vectorizamos las queries con el mismo vectorizador que usaste para el corpus
queries_vectorized = vectorizer.transform(queries_cleaned)

# Convertimos las queries vectorizadas a un DataFrame para visualizarlas
queries_df = pd.DataFrame(queries_vectorized.toarray(), columns=vectorizer.get_feature_names_out())

# Ver el resultado de las queries vectorizadas
print(queries_df)
```
